```python
import numpy as np
import matplotlib.pyplot as plt
from LoopExtrusionModel.loopExtrusionModel import Genome, Cohesin, Extruder, Simulator
```

```python
def simulations(i, time, n_extruders=5000, per_i=10, option="cohesin"):
    sim = Simulator(Genome(1), i, n_extruders, per_i, option)
    sim.simulate(time)
    sim.plot_fold(i)

def sim(option="cohesin"):
    plt.figure(figsize=(20, 20), dpi=80)
    for i in range(50):
        simulations(i, time=500, per_i=10, option=option)
    plt.show()
```

```python
sim()
```

```python
sim(option="both")
```

```python
sim(option="both_0.2")
```

```python
sim(option="both_d_0.5")
```
