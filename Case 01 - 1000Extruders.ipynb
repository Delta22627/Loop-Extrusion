```python
import numpy as np
import matplotlib.pyplot as plt
from LoopExtrusionModel.loopExtrusionModel import Genome, Cohesin, Extruder, Simulator
```

```python
def simulations(i, time, n_extruders=1000, per_i=10, option="cohesin"):
    sim = Simulator(Genome(1), i, n_extruders, per_i, option)
    sim.simulate(time)
    sim.plot()
    sim.average_domain()
    return sim.extruders_domains
```

```python
def sim_per_i(per_i, option):
    all_domains = dict()
    for i in range(10):
        extruders_domains = simulations(i, time=200, per_i=per_i, option=option)
        for key in extruders_domains:
            if all_domains.get(key) is None:
                all_domains[key] = extruders_domains[key]
            else:
                all_domains[key].extend(extruders_domains[key])
    plt.savefig(f"sim_image_1000/{option} per {per_i} Extruders.png")
    plt.show()
    print([(k ,np.mean(x)) for k, x in all_domains.items()])
    labels, data = [*zip(*all_domains.items())]
    plt.boxplot(data)
    plt.xticks(range(1, len(labels) + 1), labels)
    plt.savefig(f"sim_image_1000/{option} per {per_i} Boxplot.png")
    plt.show()
```

```python
def run(option):
    for i in range(1,11):
        sim_per_i(i*10, option)
```

```python
run("cohesin")
```

```python
run("both")
```
