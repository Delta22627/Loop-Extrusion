```python
import numpy as np
import matplotlib.pyplot as plt
from LoopExtrusionModel.loopExtrusionModel import Genome, Cohesin, Extruder, Simulator
```

```python
def simulations(i, time, n_extruders=3000, per_i=10, option="cohesin"):
    sim = Simulator(Genome(1), i, n_extruders, per_i, option)
    sim.simulate(time)
    sim.plot()
```

```python
def sim_per_i(per_i):
    for i in range(100):
        simulations(i, time=200, per_i=per_i)
    plt.show()
for i in range(1,11):
    sim_per_i(i*10)
```

```python
def sim_per_i(per_i):
    for i in range(100):
        simulations(i, time=200, per_i=per_i, option="both")
    plt.show()
for i in range(1,11):
    sim_per_i(i*10)
```
